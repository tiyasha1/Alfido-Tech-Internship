# EDA(Explanatory Data Analysis)

## Event

```python
ax=sns.countplot(x='event',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

**Highest event of this website traffic data (related to music activity) is Pageview.Basically it means maximum users view this page.**

## Date

```python
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.countplot(x='date',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

**21/08/2025 is the latest date which indicates a spike in user interest beacuse Tesher released a remix version with Jason Derulo in june 2021 but went viral on that day and also it was a saturday so entertainment purpose website was higher on that day.**

## country

```python
ax=sns.countplot(x='country',data=df.head())
for bars in ax.containers:
    ax.bar_label(bars)
```

**Saudi Arabia is the highest viewer of this music website according to the above graph.**

## Artist

```python
import warnings
warnings.filterwarnings('ignore')
ax = sns.countplot(x='artist', data=df.head(), palette='Set2')  # Try Set1, Set2, or any palette
for bars in ax.containers:
    ax.bar_label(bars)
```

**Most of the song in this website viewed by peoples are Tesher's song.**

## Album

```python
ax = sns.countplot(x='album', data=df.head(), palette='Set2')  # Try Set1, Set2, or any palette
for bars in ax.containers:
    ax.bar_label(bars)
```

**Highest viewer Album of this website is Jalebi Baby because in tiktok and instagram and all other oatforms jalebi baby was viral at this time.**

## Track	

```python
ax = sns.countplot(x='track', data=df.head(), palette='Set2')  # Try Set1, Set2, or any palette
for bars in ax.containers:
    ax.bar_label(bars)
```

**Most viewed individual song(Track) is jalebi Baby**

## city

```python
ax = sns.countplot(x='city', data=df.head(), palette='Set2')  # Try Set1, Set2, or any palette
for bars in ax.containers:
    ax.bar_label(bars)
```

**Jeddah city which is located in the western coast of Saudi Arabia along red sea is the highest player of jalebi Baby song of Tesher**

## isrc

```python
ax = sns.countplot(x='isrc', data=df.head(), palette='Set2')  # Try Set1, Set2, or any palette
for bars in ax.containers:
    ax.bar_label(bars)
```

**QZNWQ2070741 is the most popular recording code.**

## linkid

```python
ax = sns.countplot(x='linkid', data=df.head(), palette='Set2')  # Try Set1, Set2, or any palette
for bars in ax.containers:
    ax.bar_label(bars)
```

**2d896d31-97b6-4869-967b-1c5fb9cd4bb8 is most engaging links.**

# Conclusion

**The song Jalebi Baby by Tesher emerged as the central driver of traffic on the music website, particularly around August 21, 2021, with Saudi Arabia (especially Jeddah) leading in user engagement. The website traffic was dominated by pageviews, reflecting interest in content exploration. The ISRC code confirms the specific track version that contributed most. This analysis shows how a viral social media trend can influence global music engagement, especially in target cities and regions**
